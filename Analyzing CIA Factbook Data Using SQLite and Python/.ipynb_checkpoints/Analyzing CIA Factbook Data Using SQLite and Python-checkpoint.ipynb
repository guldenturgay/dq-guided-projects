## Overview of the Data

In this project we'd like to analyze CIA's Factbook data using SQLite and Python. We will see birth rates, death rates, population, population growth and population density of the world in histogram.

```python
# Import pandas as sqlite3
import pandas
import sqlite3
conn = sqlite3.connect('factbook.db')
q1 = "SELECT * FROM sqlite_master WHERE type='table' ;"
pandas.read_sql_query(q1, conn)
```

```python
# query to read the facts table
q2 = 'SELECT * FROM facts'
pandas.read_sql_query(q2,conn).head()
```

# Summary Statistics

```python
# query to min and max calculatios
q3 = 'SELECT MIN(population), MAX(population),MIN(population_growth), MAX(population_growth) FROM facts '
pandas.read_sql_query(q3, conn)
```

# Outliers

```python
# bring all the rows from factsn table where the population is min
pandas.read_sql_query('SELECT * FROM facts WHERE population = 0 ', conn)
```

```python
# bring all the rows from factsn table where the population is max
pandas.read_sql_query('SELECT * FROM facts WHERE population = 7256490011', conn)
```

Antarctica is the country with a population of 0. The reason is that there is no residency in Antarctica except scientists which are on land for their research during some certain times in a year.</p>

The maximum population of 7256490011 belongs to the World, so we understand that World is in the name column as if it is a country.

# Histograms

```python
# matplotlib and seaborn to plot
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

q4 ='''
SELECT population, population_growth, birth_rate, death_rate
FROM facts
WHERE population != (SELECT MAX(population) FROM facts ) 
and population != (SELECT MIN(population) FROM facts)'''

pandas.read_sql_query(q4 , conn).hist(ax=ax)
```

# POPULATION DENSITY

```python
# bring population density of countries
q5 = '''
SELECT name, CAST(population as float)/CAST(area_land as float) population_density 
FROM facts
ORDER BY population_density DESC LIMIT 20'''
pandas.read_sql_query(q5, conn)
```

```python
# plot histogram of population density
pandas.read_sql_query(q5, conn).plot(kind= 'hist')
```

## RATIOS OF WATER TO LAND

```python
# query for water to land ratio of countries
q6 = '''
SELECT name, CAST(area_water as float)/CAST(area_land as float) water_land_ratio 
FROM facts
ORDER BY water_land_ratio DESC'''
pandas.read_sql_query(q6, conn)
```
