## __ 1. Introduction__

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
```

```python
raw_data_count = recent_grads.shape
print(raw_data_count)
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape
print(cleaned_data_count)
```

```python
recent_grads['Sample_size'].sort_values(ascending=False)
```

## __2. Pandas, Scatter Plots__

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(10,5))
```

__There is no linear correlation between popularity of a major and median salary amongst recent grads.__

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size', figsize=(10,5))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median', figsize=(10,5))
```

__Majority of full-time employees earn between 20k-40k.__

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Sharewomen vs. unemployment', figsize=(10,5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(10,5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(10,5))
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median', figsize=(10,5))
```

__Students that majored in subjects that were majority female do not make more money. On the contrary there is a slight decrease of median salary in majors that were majority female.__
