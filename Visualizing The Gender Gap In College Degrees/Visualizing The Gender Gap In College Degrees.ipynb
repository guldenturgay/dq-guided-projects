## Introduction

In this project, we will take a look at the number of women in various majors between the years of 1968 and 2011. Then we will visualize our findings by plotting to see if some majors have more women attended or vice versa.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

#Set plot line colors to colorblind mode
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

## Comparing across all degrees

```python
#Categorize the majors
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig= plt.figure(figsize = (16,20))

# Generate first column of line charts. STEM degrees.
#Loop 6 times for 6 plots
for sp in range (0,18,3):
    # We will layout the plots in a 6 rows and 3 columns, so the indexes for stem plots will be
    #1,4,7 and so on
    cat_index= int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    
    #Remove the spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    
    #Set x,y limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    
    #Set title
    ax.set_title(stem_cats[cat_index])
    
    #Remove tick marks
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
    #Set tick values
    ax.set_yticks([0,100])
    
    #Add horizontal line at y=50 position
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3)
    
    #Aff text to the plots at the given positions
    if cat_index == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2002, 8, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2001, 10, 'Women')
        ax.tick_params(labelbottom='on')
        
        
## Generate second column of line charts. LIBERAL ARTS degrees.
for sp in range (1,16,3):
    cat_index= int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if cat_index == 0:
        ax.text(2005, 83, 'Women')
        ax.text(2002, 18, 'Men')
        
    elif cat_index == 4:
        ax.tick_params(labelbottom='on')
        
        
## Generate third column of line charts. OTHER degrees.
for sp in range (2,18,3):
    cat_index= int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = 'off')
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if cat_index == 0:
        ax.text(2005, 89, 'Women')
        ax.text(2002, 6, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 30, 'Women')
        ax.tick_params(labelbottom='on')

#Save the plot file in the same folder as the notebook file
plt.savefig('gender_degrees.png')        
plt.show()
```

## Conclusion

__For STEM degrees__, it seems like there was still a huge gap in Engineering and Computer Science degrees. On the other hand Psychology and Biology majors were more popular amongst women and in Math and Statictics and Physical Sciences there was an increase of women attendance over the years.

__For Liberal Arts degrees__, we can see that Foreign Languages, English and Art and Performance major had almost the same number of women to number of men ratio over the years and they were women majority. It was the opposite for Communications and Journalism at 70s but by the 80s it became more popular amongst women. And it seems that even though Social Sciences and History major was more popular at 70s, it had the same attendance by women and men after 2000.

__For other degrees__, majors such as Arcitecture, Business and Agriculture became more popular for women over the years and gender gap has closed. On the other hand, Health Professions, Education and Public Administration was always more popular for women over the years.
